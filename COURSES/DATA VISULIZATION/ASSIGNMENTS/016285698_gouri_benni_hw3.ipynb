```python
#q-q plot
import pylab as py
import statsmodels.api as sm
c = np.array (df2['Fare'])
c = np.random.normal(c)
sm.qqplot(c)
py.show()
```

```python
a_bins = np.linspace(0,80,20)
plt.hist(df2[df2.Sex == 'male'].Age, a_bins, alpha = 0.5, label = 'male' )
plt.hist(df2[df2.Sex == 'female'].Age, a_bins, alpha = 0.5, label = 'female')
plt.legend(loc = 'upper left')
plt.title("Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
```

```python
plt.style.use('seaborn')
df2.boxplot(column = 'Age', by = 'Survived', figsize= (6,5))
plt.title('Survival Pasenger by Age')
plt.ylabel("Age")
```

```python
plt.figure(figsize=(9,8))
sns.set(font_scale = 1.0)
ax = sns.barplot(data=df2,x=df2['Pclass'].round(),y='Fare',hue='Sex',hue_order=['female','male'],ci=0)
ax.set_xlabel("Pclass",fontsize=10)
ax.set_ylabel("Fare",fontsize=10)

ax.bar_label(ax.containers[0],fmt='%.f', label='center',fontsize=15,color='white')
ax.bar_label(ax.containers[1],fmt='%.f', label='center',fontsize=15,color='white')
plt.grid("whitegrid",axis='y')
plt.show()
```

```python
corr = df2.corr()
plt.figure(figsize = (10, 10))
sns.heatmap(corr, vmax = .9, linewidths = 0.02, square = True, annot = True, cmap = 'YlGnBu',linecolor = "black")
plt.title('Correlation between features')
plt.show()
```

```python
plt.figure(figsize = (10, 9))

plt.subplot(2, 2, 1)
sns.distplot(df2['Pclass'])

plt.subplot(2,2,2)
sns.boxplot(df2['Fare'])

plt.subplot(2, 2, 3)
sns.distplot(df2['Fare'])

plt.subplot(2, 2, 4)
sns.boxplot(df2['Age'])

plt.tight_layout()
plt.show()
```

```python
fig = sns.FacetGrid(df2, hue = "Survived", aspect = 4)
fig.map(sns.kdeplot, 'Age', shade = True)
oldest = df2['Age'].max()
fig.set(xlim = (0, oldest))
fig.add_legend()
plt.title('PLOTS')
```

```python
g= sns.lmplot(x="Pclass", y="Age",data=df2)
```

```python
plt.rc('xtick', labelsize=10) 
plt.rc('ytick', labelsize=10)
plt.figure()
fig = df2[['Fare','Age']].plot.hist(histtype= 'bar', alpha = 0.8)
plt.legend(('Died','Pclass'), fontsize = 12)
plt.xlabel('Age', fontsize = 18)
plt.show()
```
